```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error

# Filter rows with missing values
df = df.dropna(axis=0)
# Choose target and features
y = df.price
features = ['bed', 'bath', 'acre_lot', 'house_size']
X = df[features]

train_X, val_X, train_y, val_y = train_test_split(X, y,random_state = 0)

forest_model = RandomForestRegressor(random_state=1)
forest_model.fit(train_X, train_y)
melb_preds = forest_model.predict(val_X)
print(mean_absolute_error(val_y, melb_preds))
```

```python
from sklearn.linear_model import LinearRegression 
model = LinearRegression().fit(X, y) 
linear_pred = model.predict(val_X)
print(mean_absolute_error(val_y, linear_pred))
r_sq = model.score(X, y)
print('coefficient of determination:', r_sq)
```

```python
from sklearn.linear_model import Ridge 
ridge_model = Ridge(alpha=1.0)
ridge_model.fit(X, y) 
ridge_pred = ridge_model.predict(val_X)
print(mean_absolute_error(val_y, ridge_pred))
rr_sq = ridge_model.score(X, y)
print('coefficient of determination:', rr_sq)
```

```python
from sklearn.linear_model import Lasso 
Lasso_model = Lasso(alpha=1.0)
Lasso_model.fit(X, y) 
Lasso_pred = Lasso_model.predict(val_X)
print(mean_absolute_error(val_y, Lasso_pred))
lr_sq = Lasso_model.score(X, y)
print('coefficient of determination:', lr_sq)
```
